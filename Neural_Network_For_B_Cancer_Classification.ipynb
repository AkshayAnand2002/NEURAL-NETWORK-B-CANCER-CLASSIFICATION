```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

```python
dataset=sklearn.datasets.load_breast_cancer()
```

```python
print(dataset)
```

```python
#benign tumours are localized and they don't travel to other parts of body.
#malignant tumours are cancerous as they can travel to other parts of body.
```

```python
dataframe=pd.DataFrame(dataset.data,columns=dataset.feature_names)
```

```python
dataframe['label']=dataset.target
```

```python
#separating features and target
X=dataframe.drop(columns='label',axis=1)
Y=dataframe['label']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
#Standardize the data
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
X_train_std=scaler.fit_transform(X_train)
X_test_std=scaler.transform(X_test)
#X train standardized and X test standardized
```

```python
#Building the Neural Network
#It consists of input layer,hidden layers and output layer.
```

```python
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
#Keras is wrapper of tensorflow as tensorflow is its backend.
```

```python
#setting up the layers now
model=keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)),
    keras.layers.Dense(20,activation='relu'),
    keras.layers.Dense(2,activation='sigmoid')
])
#flatten() converts multidimensional data to single dimensional format.
#input shape contains 30 columns that are present in the data.
#keras.layers.Dense(20,activation='relu') is the hidden layer having 20 neurons.
#Output layer has 2 neurons and activation function is sigmoid.
#no. of neurons in output layer=no. of classes in label.
```

```python
#compiling the neural network
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
#dog-0 cat-1 horse-2  <--LABEL ENCODING
```

```python
#HERE LABELS ARE 0 AND 1 SO USED SPARSE_CATEGORICAL_CROSSENTROPY
#IF WE USE ONE HOT ENCODING SO WE USE CATEGORICAL_CROSSENTROPY
```

```python
#ONE HOT ENCODING OF LABELS
#dog=[1 0 0]
#cat=[0 1 0]
#horse=[0 0 1]
```

```python
#HERE WE HAVE JUST 2 KINDS OF LABELS SO WE CAN USE LABEL ENCODING.
```

```python
#FOR LABEL ENCODING WE USE SPARSE_CATEGORICAL_CROSSENTROPY
```

```python
#TRAINING THE NEURAL NETWORK NOW
history=model.fit(X_train_std,Y_train,validation_split=0.1,epochs=10)
```

```python
#VISUALIZING ACCURACY AND LOSS
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('MODEL ACCURACY')
plt.ylabel('ACCURACY')
plt.xlabel('EPOCHS')
plt.legend(['training data','validation data'],loc='lower right')
```

```python
#VISUALIZING ACCURACY AND LOSS
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('MODEL LOSS')
plt.ylabel('LOSS')
plt.xlabel('EPOCHS')
plt.legend(['training data','validation data'],loc='upper right')
```

```python
#ACCURACY OF MODEL ON TEST DATA
loss,accuracy=model.evaluate(X_test_std,Y_test)
print(accuracy)
```
